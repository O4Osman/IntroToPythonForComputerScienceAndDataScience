```python
"""

7.10 (Project: Two-Player, Two-Dimensional Tic-Tac-Toe) Write a script to play two- dimensional Tic-Tac-Toe between two 
human players who alternate entering their moves on the same computer. Use a 3-by-3 two-dimensional array. Each player 
indicates their moves by entering a pair of numbers representing the row and column indices of the square in which they want 
to place their mark, either an 'X' or an 'O'. When the first play- er moves, place an 'X' in the specified square. When the 
second player moves, place an 'O' in the specified square. Each move must be to an empty square. After each move, determine 
whether the game has been won and whether itâ€™s a draw.

"""
```

```python
import numpy as np
```

```python
winning_positions = {
    0: [[0, 0], [0, 1], [0, 2]],
    1: [[1, 0], [1, 1], [1, 2]],
    2: [[2, 0], [2, 1], [2, 2]],
    3: [[0, 0], [1, 0], [2, 0]],
    4: [[0, 1], [1, 1], [2, 1]],
    5: [[0, 2], [1, 2], [2, 2]],
    6: [[0, 0], [1, 1], [2, 2]],
    7: [[0, 2], [1, 1], [2, 0]],
}
```

```python
tic_tac_toe = np.array([' '] * 9).reshape(3, 3)
```

```python
tic_tac_toe
```

```python
counter = 1
mark_x = 'X'
mark_o = 'O'
```

```python
def check_position(row, column):
    """check if position is empty or not"""
    return tic_tac_toe[row][column] == ' '
```

```python
def get_user_input():
    """get a pair of numbers from user to mark the position in the game"""
    global counter
    while True:
        if counter % 2 == 0:
            print('Player 2...')
        else:
            print('Player 1...')
        user_input = input('Enter a pair of number (1,1): ')
        split_input = user_input.split(',')
        
        if ' ' in split_input:
            split_input = split_input.remove(' ')
            
        if len(split_input) == 2:
            row = int(split_input[0])
            column = int(split_input[1])
            if 0 <= row <= 2 and 0 <= column <= 2:
                if check_position(row, column):
                    return row, column
                else:
                    print('Position not available, try again.')
            else:
                print('Please enter a valid number.')
        else:
            print('Please enter a valid number.')
```

```python
def check_game_status(row, column):
    for position in winning_positions.values():
        p1 = position[0]
        p2 = position[1]
        p3 = position[2]
        if tic_tac_toe[p1[0], p1[1]] == mark_x and tic_tac_toe[p2[0], p2[1]] == mark_x and tic_tac_toe[p3[0], p3[1]] == mark_x:
            print('User with mark X won.')
            return 'END'
        elif tic_tac_toe[p1[0], p1[1]] == mark_o and tic_tac_toe[p2[0], p2[1]] == mark_o and tic_tac_toe[p3[0], p3[1]] == mark_o:
            print('User with mark O won.')
            return 'END'
        if ' ' not in tic_tac_toe:
            print('It\'s a draw.')
            return 'END'
    return 'CONTINUE'
```

```python
def start_game():
    global counter, tic_tac_toe
    while True:
        row, column = get_user_input()
        
        if counter % 2 == 0:
            tic_tac_toe[row, column] = mark_o
        else:
            tic_tac_toe[row, column] = mark_x
            
        status = check_game_status(row, column)
        
        print(tic_tac_toe)
        
        if status == 'END':
            break
        
        counter += 1
```

```python
start_game()
```

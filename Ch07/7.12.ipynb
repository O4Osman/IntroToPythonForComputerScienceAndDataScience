```python
import numpy as np
import random
```

```python
game_board = None
counter = None
mark_x = 'X'
mark_o = 'O'
player1 = None
player2 = None
current_player = None
winning_positions = []
```

```python
def get_game_board_ready():
    global game_board
    game_board = np.array(['~'] * 64).reshape(4, 4, 4)
```

```python
def check_2d_rows(array):
    """check row wise for winner"""
    global winning_positions
    temp = []
    layers, rows, columns = array.shape
    for layer in range(layers):
        for row in range(rows):
            for col in range(columns):
                temp.append((layer, row, col))
                # print(array[layer][row][col])
            winning_positions += [temp]
            temp = []
        #     print()
        # print()
```

```python
def check_2d_columns(array):
    """check column wise for winner"""
    global winning_positions
    temp = []
    layers, rows, columns = array.shape
    for layer in range(layers):
        for col in range(columns):
            for row in range(rows):
                temp.append((layer, row, col))
                # print(array[layer][row][col])
            winning_positions += [temp]
            temp = []
        #     print()
        # print()
```

```python
def check_2d_daigonal(array):
    """check diagonal of each layer for winner"""
    global winning_positions
    temp = []
    layers, rows, columns = array.shape
    
    for layer in range(layers):
        
        for row in range(rows):
            temp.append((layer, row, row))
            # print(array[layer][row][row])
            continue
        winning_positions += [temp]
        temp = []
        # print()

    temp = []
    
    for layer in range(layers):

        for row in range(rows):
            temp.append((layer, row, 3 - row))
            # print(array[layer][row][3 - row])
            continue
        winning_positions += [temp]
        temp = []
        # print()
```

```python
def check_3d_vertical(array):
    """check horizontal position of each layer for winner"""
    global winning_positions
    temp = []
    layers, rows, columns = array.shape

    for row in range(rows):
        for col in range(columns):
            for layer in range(layers):
                temp.append((layer, row, col))
                # print(array[layer][row][col])
        #     print()
            winning_positions += [temp]
            temp = []
        # print()
```

```python
def check_3d_rows(array):
    """check rows in 3d for winner"""
    global winning_positions
    temp = []
    layers, rows, columns = array.shape

    for layer in range(layers):
        for row in range(rows):
            for col in range(columns):
                temp.append((col, row, col))
                # print(array[col][row][col])
        #     print()
            winning_positions += [temp]
            temp = []
        # print()
```

```python
def check_3d_columns(array):
    """check columns in 3d for winner"""
    global winning_positions
    temp = []
    layers, rows, columns = array.shape

    for layer in range(layers):
        for row in range(rows):
            for col in range(columns):
                temp.append((col, col, layer))
        #         print(array[col][col][layer])
            winning_positions += [temp]
            temp = []
        #     print()
        # print()
```

```python
def check_3d_daigonal(array):
    """check daigonals in 3d for winner"""
    global winning_positions
    temp = []
    layers, rows, columns = array.shape

    for layer in range(layers):
        for row in range(rows):
            temp.append((row, row, row))
            # print(array[row][row][row])
        winning_positions += [temp]
        temp = []
        # print()
    
    temp = []
    for layer in range(layers):
        for row in range(rows):
            temp.append((row, row, 3 - row))
        #     print(array[row][row][3 - row])
        # print()
        winning_positions += [temp]
        temp = []
```

```python
def get_all_winning_positions():
    """get and add all winning positions in winning position list"""
    global game_board

    check_2d_rows(game_board)
    check_2d_columns(game_board)
    check_2d_daigonal(game_board)
    check_3d_vertical(game_board)
    check_3d_rows(game_board)
    check_3d_columns(game_board)
    check_3d_daigonal(game_board)
```

```python
def check_position(layer, row, column):
    global game_board
    """check if position is avvailable or not"""
    return game_board[layer][row][column] == '~'
```

```python
def select_player():
    """ask user if he wants to play first"""
    global player1, player2
    user_input = input('Do you want to play first (Y/N): ').upper()
    
    if user_input == 'Y':
        player1 = 'User'
        player2 = 'Computer'
    elif user_input == 'N':
        player1 = 'Computer'
        player2 = 'User'
    else:
        print('Enter correct response.')
```

```python
def get_computer_input():
    """get a pair of numbers from computer to mark the position in the game"""
    while True:
        layer = random.randrange(0, 4)
        row = random.randrange(0, 4)
        column = random.randrange(0, 4)

        if check_position(layer, row, column):
            print(f"Computer chose position ({layer}, {row}, {column})")
            return layer, row, column
```

```python
def get_user_input():
    """get a pair of numbers from user to mark the position in the game"""
    global counter
    
    while True:
        user_input = input('Enter a pair of number (1,1,1): ')
        split_input = user_input.split(',')
        
        for _ in range(split_input.count(' ')):
            split_input = split_input.remove(' ')
            
        if len(split_input) == 3:
            layer = int(split_input[0])
            row = int(split_input[1])
            column = int(split_input[2])
            
            if 0 <= layer <= 3 and 0 <= row <= 3 and 0 <= column <= 3:
                if check_position(layer, row, column):
                    print(f'You choose ({layer}, {row}, {column})')
                    return layer, row, column
                else:
                    print('Position not available, try again.')
            else:
                print('Please enter a valid number.')
        else:
            print('Please enter a valid number.')
```

```python
def player_turn():
    """it check for the player who will play now"""
    global player1, player2, current_player
    if counter % 2 == 0:
        print(f'Player :  {player2}')
        current_player = player2
    else:
        print(f'Player : {player1}')
        current_player = player1
        
    if current_player == 'User':
        layer, row, column = get_user_input()
    else:
        layer, row, column = get_computer_input()
    return layer, row, column
```

```python
def check_game_status():
    """check if game is won by anyone or if its a draw"""
    global game_board
    
    for pos in winning_positions:
        p1 = pos[0][0]
        p2 = pos[0][1]
        p3 = pos[0][2]
        p4 = pos[1][0]
        p5 = pos[1][1]
        p6 = pos[1][2]
        p7 = pos[2][0]
        p8 = pos[2][1]
        p9 = pos[2][2]
        p10 = pos[3][0]
        p11 = pos[3][1]
        p12 = pos[3][2]
        
        if game_board[p1,p2,p3] == game_board[p4,p5,p6] == game_board[p7,p8,p9] == game_board[p10,p11,p12] and game_board[p1,p2,p3] != '~':
            sign = game_board[p1,p2,p3]
            if sign == 'X':
                print(f'{player1} Won.')
                return  'WON'
            elif sign == 'O':
                print(f'{player2} Won.')
                return 'WON'
    if '~' not in game_board:
        print('Its a draw.')
        return 'Draw'
    return 'CONTINUE'
```

```python
def start_game():
    """this function initialize all other requried functions to set up the game and run the game"""
    global counter, game_board, mark_x, mark_o
    get_game_board_ready()
    counter = 1
    get_all_winning_positions()
    select_player()
    while True:
        layer, row, column = player_turn()
        if counter % 2 == 0:
            game_board[layer, row, column] = mark_o
        else:
            game_board[layer, row, column] = mark_x
        print(game_board)
        status = check_game_status()
        if status != 'CONTINUE':
            break
        counter += 1
```

```python
start_game()
```

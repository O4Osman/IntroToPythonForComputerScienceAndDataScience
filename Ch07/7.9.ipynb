```python
"""

7.9 (Indexing and Slicing arrays) Create an array containing the values 1–15, reshape it into a 3-by-5 array, then use 
indexing and slicing techniques to perform each of the following operations:
a) Select row 2.
b) Select column 5.
c) Select rows 0 and 1.
d) Select columns 2–4.
e) Select the element that is in row 1 and column 4.
f) Select all elements from rows 1 and 2 that are in columns 0, 2 and 4.

"""
```

```python
import numpy as np
```

```python
arr = np.arange(1, 16).reshape(3, 5)
```

```python
arr
```

```python
# a) Select row 2.
arr[[2], : ]
```

```python
# b) Select column 5.
arr[:, [4]]
```

```python
# c) Select rows 0 and 1.
arr[[0, 1], : ]
```

```python
# d) Select columns 2–4.
arr[:, 2:5]
```

```python
# e) Select the element that is in row 1 and column 4.
arr[1, 4]
```

```python
# f) Select all elements from rows 1 and 2 that are in columns 0, 2 and 4.
arr[1:3 , [0, 2, 4]]
```

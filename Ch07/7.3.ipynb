```python
"""

7.3 (Element-Wise array Multiplication) Create a 3-by-3 array containing the even integers from 2 through 18. Create a 
second 3-by-3 array containing the integers from 9 down to 1, then multiply the first array by the second.

"""
```

```python
import numpy as np
```

```python
arr1 = np.arange(2, 19, 2).reshape(3, 3)
```

```python
arr2 = np.arange(9, 0, -1).reshape(3, 3)
```

```python
arr3  = arr1 * arr2
```

```python
import numpy as np
import random
```

```python
# a) 
board = None
horizontal = np.array([2, 1, -1, -2, -2, -1, 1, 2])
vertical = np.array([-1, -2, -2, -1, 1, 2, 2, 1])
current_row = None
current_column = None
counter = None
```

```python
def initialize_game(row, column):
    """reset board and current_row and column and counter"""
    global board, current_row, current_column, counter
    board = np.array([0] * 64).reshape(8, 8)
    current_row = row
    current_column = column
    counter = 63
```

```python
initialize_game(0, 0)
while True:
    old_row = current_row
    old_column = current_column
    possible_moves = []
    move_number = random.randrange(8)
    for move_number in range(8):
        possible_row = current_row + vertical[move_number]
        possible_column = current_column + horizontal[move_number]
        if 0 <= possible_row <= 7 and 0 <= possible_column <= 7:
            if board[possible_row][possible_column] == 0:
                possible_moves += [move_number]
                
    if possible_moves:
        random_number = random.randrange(len(possible_moves))
        selected_move = possible_moves[random_number]
        
        counter += 1
        board[current_row][current_column] = counter
        current_row += vertical[selected_move] 
        current_column += horizontal[selected_move]
    
    if old_row == current_row and old_column == current_column:
        counter += 1
        board[current_row][current_column] = counter
        break
        
print(f'Maximum Moves : {counter}')
print(board)
```

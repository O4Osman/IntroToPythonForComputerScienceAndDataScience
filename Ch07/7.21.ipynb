```python
"""

7.21 (Shallow vs. Deep Copy) In this chapter, we discussed shallow vs. deep copies of arrays. Python’s built-in list and 
dictionary types have copy methods that perform shallow copies. Using the following dictionary
       dictionary = {'Sophia': [97, 88]}
demonstrate that a dictionary’s copy method indeed performs a shallow copy. To do so, call copy to make the shallow copy, 
modify the list stored in the original dictionary, then display both dictionaries to see that they have the same contents.
Next, use the copy module’s deepcopy function to create a deep copy of the dictionary. Modify the list stored in the 
original dictionary, then display both dictionaries to prove that each has its own data.

"""
```

```python
from copy import deepcopy
```

```python
# Original dictionary
dictionary = {'Sophia': [97, 88]}
dictionary
```

```python
# copy of the dictionary
dictionary_copy = dictionary.copy()
dictionary_copy
```

```python
# modifying the original dictionary
dictionary['Sophia'] += [85]
```

```python
# original dictionary
dictionary
```

```python
# copied dictionary
dictionary_copy
```

```python
# so copies dictionary is pointing to original dictionary. So copy() perform shallow copy
```

```python
# making a deep copy
deep_copy = deepcopy(dictionary)
```

```python
# modify the original dictionary
dictionary['Sophia'] += [100]
```

```python
# now original dictionary
dictionary
```

```python
# deep copied dictionary
deep_copy
```

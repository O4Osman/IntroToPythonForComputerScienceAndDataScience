```python
"""

7.19 (Enhanced Median and Mode of an array) Modify your functions from the pre- vious exercise to allow the user to provide 
an axis keyword argument so the calculations can be performed row-by-row or column-by-column on a two-dimensional array.

"""
```

```python
import numpy as np
import math
```

```python
def array_median(array):
    """find and return median of the array"""
    rows, columns = array.shape
    straight_array = sorted(np.ravel(array))
    length = (len(straight_array))
    if length % 2 == 0:
        first_element = math.floor(length / 2)
        median = (straight_array[first_element - 1] + straight_array[first_element]) / 2
        return median
    else:
        center_element = math.ceil(length / 2)
        return straight_array[center_element - 1]
```

```python
def array_mode(array):
    """find and return mode of the array"""
    values, frequencies = np.unique(array, return_counts=True)
    max_index = None
    maximum = frequencies.max()
    for index, val in enumerate(frequencies):
        if val == maximum:
            max_index = index
            break
    
    return values[max_index]
```

```python
def get_input():
    while True:
        user_input = int(input('Enter axis (0/1): '))
        if user_input not in [0, 1]:
            print('Enter valid number.')
        else:
            return user_input
```

```python
def get_median(array, axis):
    """this function get median according to axis"""
    rows, columns = array.shape
    median_list = []
    if axis == 1:
        for row in range(rows):
            median_list += [array_median(array[[row],:])]
    elif axis == 0:
        for col in range(columns):
            median_list += [array_median(array[:, [col]])]
    return np.array(median_list)
```

```python
def get_mode(array, axis):
    """this function get mode according to axis"""
    rows, columns = array.shape
    mode_list = []
    if axis == 1:
        for row in range(rows):
            mode_list += [array_mode(array[[row],:])]
    elif axis == 0:
        for col in range(columns):
            mode_list += [array_mode(array[:, [col]])]
    return np.array(mode_list)
```

```python
sample = [[10, 20, 30, 40], [10, 50, 60, 70], [10, 80, 90, 100]]
```

```python
arr = np.array(sample)
axis = get_input()
```

```python
median = get_median(arr, axis)
mode = get_mode(arr, axis)
print(f'Median : {median}')
print(f'Mode: {mode}')
```

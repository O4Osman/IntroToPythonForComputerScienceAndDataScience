```python
import pandas as pd
```

```python
def get_character_lenght(number):
    """convert number into string and return character length of that number"""
    return len(f'{number}')
```

```python
def data_frame_formatting(dt):
    """reimplement pandas's output function using loops to understand built-in output function"""

    rows, columns = dt.shape

    row_index_width = len(f'{rows - 1}')

    # get max widths of columns
    column_widths = []
    for col in range(columns):
        max_number = dt.iloc[:,[col]].max()
        width = get_character_lenght(max_number.iloc[0])
        column_widths += [width + 2]
        
    print(f'{' ':>{row_index_width}}', end=' ')

    # print columns heading
    for col in range(columns):
        w = column_widths[col]
        print(f'{col:>{w}}', end='')
    
    # print dataframe
    for row in range(rows):
        print(f'\n{row:>{row_index_width}}', end=' ')
        
        for col in range(columns):
            w = column_widths[col]
            print(f'{dt.iloc[row, col]:>{w}}', end='')
```

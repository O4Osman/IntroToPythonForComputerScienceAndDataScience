```python
"""

7.22 (Pandas: Series) Perform the following tasks with pandas Series: 

a) Create a Series from the list [7, 11, 13, 17].
b) Create a Series with five elements that are all 100.0.
c) Create a Series with 20 elements that are all random numbers in the range 0 to 100. Use method describe to produce the 
Series’ basic descriptive statistics.
d) Create a Series called temperatures of the floating-point values 98.6, 98.9, 100.2 and 97.9. Using the index keyword 
argument, specify the custom indices 'Julie', 'Charlie', 'Sam' and 'Andrea'.
e) Form a dictionary from the names and values in Part(d),then use it to initialize a Series.

"""
```

```python
import numpy as np
import pandas as pd
import random
```

```python
# Create a Series from the list [7, 11, 13, 17].
series = pd.Series([7, 11, 13, 17])
series
```

```python
# Create a Series with five elements that are all 100.0.
series1 = pd.Series(100.0 , range(5))
series1
```

```python
# c) Create a Series with 20 elements that are all random numbers in the range 0 to 100. Use method describe to produce the 
# Series’ basic descriptive statistics.

series2 = pd.Series(np.random.randint(0, 100, 20))
print(series2)
series2.describe()
```

```python
# d) Create a Series called temperatures of the floating-point values 98.6, 98.9, 100.2 and 97.9. Using the index keyword 
# argument, specify the custom indices 'Julie', 'Charlie', 'Sam' and 'Andrea'.
series3 = pd.Series([98.6, 98.9, 100.2, 97.9], index=['Julie', 'Charlie', 'Sam', 'Andrea'])
series3
```

```python
# Form a dictionary from the names and values in Part(d),then use it to initialize a Series.
dictionary = {
    'Julie': 98.6,
    'Charlie': 98.9,
    'Sam': 100.2,
    'Andrea': 97.9
}
series4 = pd.Series(dictionary)
series4
```

```python
"""

7.23 (Pandas: DataFrames) Perform the following tasks with pandas DataFrames:
a) Create a DataFrame named temperatures from a dictionary of three temperature readings each for 'Maxine', 'James' and 
'Amanda'.
b) Recreate the DataFrame temperatures in Part (a) with custom indices using the index keyword argument and a list 
containing 'Morning', 'Afternoon' and 'Evening'.
c) Select from temperatures the column of temperature readings for 'Maxine'.
d) Select from temperatures the row of 'Morning' temperature readings.
e) Select from temperatures the rows for 'Morning' and 'Evening' temperature readings.
f) Select from temperatures the columns of temperature readings for 'Amanda' and 'Maxine'.
g) Select from temperatures the elements for 'Amanda' and 'Maxine' in the 'Morning' and 'Afternoon'.
h) Use the describe method to produce temperatures’ descriptive statistics.
i) Transpose temperatures.
j) Sort temperatures so that its column names are in alphabetical order.

"""
```

```python
import pandas as pd
```

```python
# a) Create a DataFrame named temperatures from a dictionary of three temperature readings each for 'Maxine', 'James' and 
# 'Amanda'.

dictionary = {
    'Maxine': [50, 51, 52],
    'James': [54, 55, 56],
    'Amanda': [49, 50, 51],
}

temperatures = pd.DataFrame(dictionary)
temperatures
```

```python
# b) Recreate the DataFrame temperatures in Part (a) with custom indices using the index keyword argument and a list 
# containing 'Morning', 'Afternoon' and 'Evening'.

dictionary = {
    'Maxine': [50, 51, 52],
    'James': [54, 55, 56],
    'Amanda': [49, 50, 51],
}

temperatures = pd.DataFrame(dictionary, index=['Morning', 'Afternoon', 'Evening'])
temperatures
```

```python
# c) Select from temperatures the column of temperature readings for 'Maxine'.
temperatures.Maxine
```

```python
# d) Select from temperatures the row of 'Morning' temperature readings.

temperatures.loc['Morning']
```

```python
# e) Select from temperatures the rows for 'Morning' and 'Evening' temperature readings.

temperatures.loc[['Morning','Evening'], :]
```

```python
# f) Select from temperatures the columns of temperature readings for 'Amanda' and 'Maxine'.

temperatures.loc[:, ['Amanda','Maxine']]
```

```python
# g) Select from temperatures the elements for 'Amanda' and 'Maxine' in the 'Morning' and 'Afternoon'.
temperatures.loc[['Morning','Afternoon'], ['Amanda','Maxine']]
```

```python
# h) Use the describe method to produce temperatures’ descriptive statistics.

temperatures.describe()
```

```python
# i) Transpose temperatures.

temperatures.T
```

```python
# j) Sort temperatures so that its column names are in alphabetical order.

temperatures.sort_index(axis=1)
```

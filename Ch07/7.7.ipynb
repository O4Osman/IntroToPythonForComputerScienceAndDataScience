```python
"""

7.7 (Challenge Project: Reimplement NumPy array Output) You saw that NumPy outputs two-dimensional arrays in a nice 
column-based format that right-aligns every element in a field width. The field width’s size is determined by the array 
element value that requires the most character positions to display. To understand how powerful it is to have this 
formatting simply built-in, write a function that reimplements NumPy’s array formatting for two-dimensional arrays using 
loops. Assume the array contains only positive integer values.

"""
```

```python
import numpy as np
```

```python
def array_formatting(arr):
    """reimplement array's output function using loops to understand built-in output function"""
    max_number = arr.max()
    max_width = len(f'{max_number}')
    
    print('array([', end='')
    
    for index1, row in enumerate(arr):
        print('[', end='')
        
        for index, col in enumerate(row):
            print(f'{col:>{max_width}}', end='')
            if index != (len(row) - 1):
                print( end=', ')
                
        print(']', end='')
        if index1 != (len(arr) - 1):
            print(',')
            print(f'{'':>{max_width+2}}', end='')
        if index1 == (len(arr) - 1):
            print('])')
```

```python
arr = np.array([[1, 2, 3],[2000, 30000, 500000]]).reshape(2, 3)
```

```python
arr
```

```python
array_formatting(arr)
```

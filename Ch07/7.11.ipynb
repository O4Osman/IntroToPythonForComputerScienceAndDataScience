```python
import numpy as np
import random
```

```python
winning_positions = {
    0: [[0, 0], [0, 1], [0, 2]],
    1: [[1, 0], [1, 1], [1, 2]],
    2: [[2, 0], [2, 1], [2, 2]],
    3: [[0, 0], [1, 0], [2, 0]],
    4: [[0, 1], [1, 1], [2, 1]],
    5: [[0, 2], [1, 2], [2, 2]],
    6: [[0, 0], [1, 1], [2, 2]],
    7: [[0, 2], [1, 1], [2, 0]],
}
```

```python
tic_tac_toe = None
```

```python
def create_array():
    """define the empty array"""
    global tic_tac_toe
    tic_tac_toe = np.array([' '] * 9).reshape(3, 3)
```

```python
counter = None
mark_x = 'X'
mark_o = 'O'
player1 = None
player2 = None
current_player = None
```

```python
def check_position(row, column):
    """check if position is empty or not"""
    return tic_tac_toe[row][column] == ' '
```

```python
def get_computer_input():
    """get a pair of numbers from computer to mark the position in the game"""
    while True:
        row = random.randrange(0, 3)
        column = random.randrange(0, 3)

        if check_position(row, column):
            print(f"Computer chose position ({row}, {column})")
            return row, column
```

```python
def get_user_input():
    """get a pair of numbers from user to mark the position in the game"""
    global counter
    while True:
        user_input = input('Enter a pair of number (1,1): ')
        split_input = user_input.split(',')
        
        if ' ' in split_input:
            split_input = split_input.remove(' ')
            
        if len(split_input) == 2:
            row = int(split_input[0])
            column = int(split_input[1])
            if 0 <= row <= 2 and 0 <= column <= 2:
                if check_position(row, column):
                    return row, column
                else:
                    print('Position not available, try again.')
            else:
                print('Please enter a valid number.')
        else:
            print('Please enter a valid number.')
```

```python
def check_game_status(row, column):
    for position in winning_positions.values():
        p1 = position[0]
        p2 = position[1]
        p3 = position[2]
        if tic_tac_toe[p1[0], p1[1]] == mark_x and tic_tac_toe[p2[0], p2[1]] == mark_x and tic_tac_toe[p3[0], p3[1]] == mark_x:
            print(f'{current_player} won.')
            return 'END'
        elif tic_tac_toe[p1[0], p1[1]] == mark_o and tic_tac_toe[p2[0], p2[1]] == mark_o and tic_tac_toe[p3[0], p3[1]] == mark_o:
            print(f'{current_player} won.')
            return 'END'
    if ' ' not in tic_tac_toe:
            print('It\'s a draw.')
            return 'END'
    return 'CONTINUE'
```

```python
def who_first():
    global player1, player2, current_player
    while True:
        answer = input('Do you want to play first (Y/N): ').upper()
        if answer == 'Y':
            player1 = 'User'
            player2 = 'Computer'
        elif answer == 'N':
            player1 = 'Computer'
            player2 = 'User'
        else:
            print('Enter correct response.')
```

```python
def get_positions():
    global player1, player2, current_player
    if counter % 2 == 0:
        print(f'Player :  {player2}')
        current_player = player2
    else:
        print(f'Player : {player1}')
        current_player = player1
        
    if current_player == 'User':
        row, column = get_user_input()
    else:
        row, column = get_computer_input()
    return row, column
```

```python
def start_game():
    """start the game by calling the functions"""
    global counter, tic_tac_toe
    counter = 1
    create_array()
    who_first()
    while True:
        row, column = get_positions()
        
        if counter % 2 == 0:
            tic_tac_toe[row, column] = mark_o
        else:
            tic_tac_toe[row, column] = mark_x
            
        status = check_game_status(row, column)
        
        print(tic_tac_toe)
        
        if status == 'END':
            break
        
        counter += 1
```

```python
start_game()
```

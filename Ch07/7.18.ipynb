```python
"""

7.18 (Median and Mode of an array) NumPy arrays offer a mean method, but not median or mode. Write functions median and mode
that use existing NumPy capabilities to determine the median (middle) and mode (most frequent) of the values in an array. 
Your functions should determine the median and mode regardless of the arrayâ€™s shape. Test your function on three arrays of 
different shapes.

"""
```

```python
import numpy as np
import math
```

```python
def array_median(array):
    """find and return median of the array"""
    straight_array = sorted(np.ravel(array))
    length = (len(straight_array))
    if length % 2 == 0:
        first_element = math.floor(length / 2)
        median = (straight_array[first_element - 1] + straight_array[first_element]) / 2
        return median
    else:
        center_element = math.ceil(length / 2)
        return straight_array[center_element - 1]
```

```python
def array_mode(array):
    """find and return mode of the array"""
    values, frequencies = np.unique(array, return_counts=True)
    max_index = None
    maximum = frequencies.max()
    for index, val in enumerate(frequencies):
        if val == maximum:
            max_index = index
            break
    
    return values[max_index]
```

```python
sample1 = [1, 2, 3, 4, 5, 5, 6, 7, 8, 9]

sample3 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
sample4 = [1.1, 2.2, 3.3, 2.2, 4.4]
```

```python
arr = np.array(sample1)
print(f'Median : {array_median(arr)}')
print(f'Mode : {array_mode(arr)}')
```

```python
arr = np.array(sample2)
print(f'Median : {array_median(arr)}')
print(f'Mode : {array_mode(arr)}')
```

```python
arr = np.array(sample3)
print(f'Median : {array_median(arr)}')
print(f'Mode : {array_mode(arr)}')
```

```python
arr = np.array(sample4)
print(f'Median : {array_median(arr)}')
print(f'Mode : {array_mode(arr)}')
```

```python
"""

8.17 (Regular Expressions: Matching Numeric Values) Write a regular expression that searches a string and matches a valid 
number. A number can have any number of digits, but it can have only digits and a decimal point and possibly a leading sign. 
The decimal point is optional, but if it appears in the number, there must be only one, and it must have digits on its left 
and its right. There should be whitespace or a beginning or end-of-line character on either side of a valid number.

"""
```

```python
import re
```

```python
numbers = """
+123 
 -45.67
+0.89
.56
78.
+123.456
-0.0
abc123
 123 abc
  456  
789.01 end
start 234..56 
 -12.3&
"""
```

```python
pattern = r'(^|\s)([+-]?\d+\.\d+)(\s|$)|(^|\s)([+-]?\d+)(\s|$)' # I could not solve this so I used claud.ai

for match in re.finditer(pattern, numbers):
    if match.group(2):  # Decimal number matched
        print(match.group(2))
    elif match.group(5):  # Integer matched
        print(match.group(5))
```

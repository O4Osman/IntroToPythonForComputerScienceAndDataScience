```python
def get_user_input():
    """get input from user and return list of quantity and ingredient from that user input"""
    while True:
        text = input('Enter recipe: ')
        result = re.findall(r'(\d+)\s([\w+ _-]+)|(\d+\.\d+)\s([\w+ _-]+)', text)
        if len(result) > 0:
            for index, item in enumerate(result):
                t = ()
                for x in item:
                    if x != '':
                        t += (x,)
                result[index] = t
            print(result)
            return result
        else:
            print('Invalid input. Please try again')
```

```python
def get_alternative(quantity, ingredient):
    """return alternative ingredients if that is present in dictionary"""
    if ingredient not in ingredient_substitutes:
        return f'No substitutes found for {quantity} {ingredient}'
        
    result = ''
    alternatives_list = ingredient_substitutes[ingredient]
    
    for index, dictionary in enumerate(alternatives_list):
        
        for index1, item in enumerate(dictionary['components']):
            result += f'{item['amount'] * quantity}'
            result += ' ' + item['ingredient']
            if index1 != (len(dictionary['components']) - 1):
                result += ' and '    
                
        if index != (len(alternatives_list) - 1):
            result += ' or '
    return result
```

```python
def process_list(result):
    """process the ingredients and get their alternatives"""
    result_list = []

    for index, item in enumerate(result):
        pattern = re.sub(' ','? ', item[1])
        pattern = pattern + '?'

        for ingredient in ingredient_substitutes.keys():
            if re.fullmatch(f'{pattern}', ingredient):
                text = item[0] + ' ' + item[1] + ' => '
                text += get_alternative(float(item[0]), ingredient)
                result_list += [text]
    return result_list
```

```python
def print_result(result_list):
    """print a healthcare message and alternatives of ingredients"""
    print('\n\nAlways consult your healthcare professional before making significant changes to your diet.\n')
    if len(result_list) > 0:
        print('Following is the list of ingredients that can be replaced.')
        for item in result_list:
            print(item)
    else:
        print("Sorry! we don't have any suggestion for current recipe")
```

```python
result = get_user_input()
result_list = process_list(result)
print_result(result_list)
```

```python
"""

8.10 (Project: Simple Sentiment Analysis) Search online for lists of positive sentiment words and negative sentiment words. 
Create a script that inputs text, then determines whether that text is positive or negative, based on the total number of 
positive words and the total number of negative words. Test your script by searching for Twitter tweets on a topic of your 
choosing, then entering the text for several tweets. In the data science case study chapters, weâ€™ll take a deeper look at 
sentiment analysis.

"""
```

```python
positive_words = {
    "admire", "amazing", "awesome", "beautiful", "best", "bliss", "brave", "brilliant", "calm", "cheerful",
    "confident", "cool", "creative", "delightful", "determined", "eager", "easy", "elegant", "energetic",
    "enjoyable", "excellent", "excited", "fantastic", "fine", "friendly", "funny", "generous", "gentle",
    "glad", "good", "great", "happy", "healthy", "helpful", "honest", "hopeful", "incredible", "inspiring",
    "joyful", "kind", "lively", "lovely", "lucky", "magnificent", "marvelous", "nice", "optimistic",
    "outstanding", "peaceful", "perfect", "pleasant", "positive", "powerful", "proud", "remarkable",
    "resilient", "romantic", "safe", "secure", "smart", "strong", "successful", "superb", "sweet",
    "talented", "terrific", "thankful", "thoughtful", "thriving", "trustful", "truthful", "unique",
    "upbeat", "vibrant", "victorious", "warm", "welcoming", "wonderful", "worthy"
}

negative_words = {
    "angry", "annoying", "awful", "bad", "boring", "broken", "cold", "confused", "cruel", "damaged",
    "dark", "dead", "depressing", "dirty", "disgusting", "dreadful", "embarrassed", "envious", "evil",
    "failure", "fake", "fearful", "frustrated", "greedy", "grim", "guilty", "hard", "hate", "horrible",
    "hopeless", "hurt", "ignorant", "impatient", "impolite", "incompetent", "incomplete", "inferior",
    "jealous", "lazy", "lonely", "lost", "mad", "messy", "miserable", "nasty", "negative", "nervous",
    "poor", "powerless", "regret", "rejected", "rude", "sad", "scared", "selfish", "sick", "slow",
    "sorrow", "stressful", "stupid", "terrible", "tense", "tired", "ugly", "unhappy", "upset", "useless",
    "weak", "worried", "worthless", "wrong", 'broke', 'worst'
}
```

```python
positive_count = None
negative_count = None
def reset():
    """initializes variables to 0"""
    global positive_count, negative_count
    positive_count = 0
    negative_count = 0
```

```python
def check_positive(tweet):
    """check for positive sentiments"""
    global positive_words, positive_count
    for word in tweet:
        if word.lower() in positive_words:
            positive_count += 1
```

```python
def check_negative(tweet):
    """check for negative sentiments"""
    global negative_words, negative_count
    for word in tweet:
        if word.lower() in negative_words:
            negative_count += 1
```

```python
def clean_string(tweet):
    """cleaning the string"""
    if '!' in tweet:
        tweet = tweet.replace('!', '')
    if '@' in tweet:
        tweet = tweet.replace('@', '')
    if '#' in tweet:
        tweet = tweet.replace('#', '')
    return tweet
```

```python
def start():
    """handle main logic"""
    global positive_count, negative_count
    while True:
        user_in = input('Enter a tweet: ')
        if len(user_in) > 0:
            break
        else:
            print('Invalid input.')
    clean_input = clean_string(user_in)
    split_input = clean_input.split()
    reset()
    check_positive(split_input)
    check_negative(split_input)
    if positive_count > negative_count:
        print(f"It's a positive tweet. (Positive: {positive_count}, Negative: {negative_count})")
    elif positive_count < negative_count:
        print(f"It's a nagative tweet. (Positive: {positive_count}, Negative: {negative_count})")
    elif positive_count == negative_count and positive_count != 0:
        print("It's a balanced tweet.")
    else:
        print('No result')
```

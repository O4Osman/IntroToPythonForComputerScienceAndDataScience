```python
"""

8.14 (Regular Expressions: Capturing Substrings) Reimplement Exercises 8.5 and 8.6 using regular expressions that capture 
the matching substrings, then display them.

"""
```

```python
import re
```

```python
# 8.5

while True:
    user_in = input('Enter a line: ')
    if len(user_in) > 0:
        break
    else:
        print('Invalid input.')
        
s_input = user_in.split()

results = []

for word in s_input:
    results += re.findall(r'b[a-z]+', word, flags=re.IGNORECASE)

for w in results:
    print(w)
```

```python
# 8.6

while True:
    user_in = input('Enter a line: ')
    if len(user_in) > 0:
        break
    else:
        print('Invalid input.')
        
s_input = user_in.split()

results = []

for word in s_input:
    results += re.findall(r'[a-z]+ed', word, flags=re.IGNORECASE)

for w in results:
    print(w)
```

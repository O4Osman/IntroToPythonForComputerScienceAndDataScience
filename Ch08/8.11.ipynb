```python
"""

8.11 (Project: Evaluate Word Problems) Write a script that enables the user to enter mathematical word problems like 
“two times three” and “seven minus five”, then use string processing to break apart the string into the numbers and the 
operation and return the result. So “two times three” would return 6 and “seven minus five” would return 2. To keep things 
simple, assume the user enters only the words for the numbers 0 through 9 and only the operations 'plus', 'minus', 'times' 
and 'divided by'.

"""
```

```python
digits = {
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
}

operations = {
    'plus': '+',
    'minus': '-',
    'times': '*',
    'divided by': '/'
}
```

```python
def get_user_input():
    """get input from user and evaluate it"""
    while True:
        user_in = input('Enter a math problem: ')
        s_input = user_in.split()

        if len(s_input) == 4:
            s_input[1] = s_input[1] + ' ' + s_input[2]
            del s_input[2]
        
        if len(s_input) == 3:
            if s_input[0] in digits and s_input[2] in digits and s_input[1] in operations:
                return s_input
            else:
                print('Value out of range or operation out of scope.')
        else:
            print('Invalid input.')
```

```python
def calculate(user_input):
    """calculate the result"""
    operand1 = digits[user_input[0]]
    operand2 = digits[user_input[2]]
    operator = operations[user_input[1]]

    if operand2 == 0 and operator == '/':
        return 'Error: Division by zero'

    return eval(f'{operand1} {operator} {operand2}')
```

```python
user_input = get_user_input()
calculate(user_input)
```

```python
"""

8.5 (Tokenizing and Comparing Strings) Write a script that reads a line of text, tokenizes the line using space characters 
as delimiters and outputs only those words beginning with the letter 'b'.

"""
```

```python
while True:
    user_in = input('Enter a line: ')
    if len(user_in) > 0:
        break
    else:
        print('Invalid input.')
```

```python
words = user_in.split()
```

```python
b_words = [word for word in words if word.startswith('b')]
```

```python
for w in b_words:
    print(w)
```

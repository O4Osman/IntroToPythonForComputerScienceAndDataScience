```python
"""

8.18 (Regular Expression: Password Format Validator) Search online for secure password recommendations, then research 
existing regular expressions that validate secure passwords. Two examples of password requirements are:

• Passwords must contain at least five words, each separated by a hyphen, a space, a period, a comma or an underscore.

• Passwords must have a minimum of 8 characters and contain at least one each from uppercase characters, lowercase characters,
digits and punctuation characters (such as characters in '!@#$%<^>&*?').

Write regular expressions for each of the two requirements above, then use them to test sample passwords.

"""
```

```python
import re
```

```python
password1 = """
apple-banana-mango-grape-peach
apple-banana-mango-grape
apple-banana-mango-grape-
applebanana-mango-grape-peach
sky sun.moon,star_cloud-fox
tree.house.flower.grass.river
lion,tiger,bear,wolf,fox
apple  banana  mango  grape  peach
dream_light_magic_star_rain
data-byte-code-bug-script
fast,smart,kind,brave,strong
pple_banana,mango.grape peach
"""

password2 = """
Abc123!@
Passw0rd!
abcdefghi
ABCDEFGH
MySecure9*
A1b2C3d4?
Hello123#
Abcdefgh
Abc12345
GoodLuck7&
ABC123!@
CodeTest8^
abc123!@
Strong9@Mix
Qa1Zx2C!
Ab1!
"""
```

```python
pattern = r'((\w+[-,. _]){4,}\w+)'
for match in re.finditer(pattern, password1):
    print(match.group(1))
```

```python
pattern1 = r'((?=.*[A-Z])(?=.*[a-z])(?=.*[0-9])(?=.*[!@#$%<^>&*?]).{8,})'
for match in re.finditer(pattern1, password2):
    print(match.group(1))
```

```python
"""

8.4 (Reversing a Sentence) Write a script that reads a line of text as a string, tokenizes the string with the split method 
and outputs the tokens in reverse order. Use space characters as delimiters.

"""
```

```python
while True:
    user_in = input('Enter a line: ')
    if len(user_in) > 0:
        break
    else:
        print('Invalid input.')
```

```python
splited_input = user_in.split()
```

```python
reversed_input = reversed(splited_input)
```

```python
print(' '.join(word for word in reversed_input))
```

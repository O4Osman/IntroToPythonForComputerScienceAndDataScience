```python
import re
```

```python
def search(regular_expression, text):
    """this return result of search of regular-expressions"""
    result = re.search(regular_expression, text)
    if result:
        word = result.group()
        span = result.span()
        print(f'Word: {word}')
        print(f'Index location: {span}')
    else:
        print('No match found.')
```

```python
exp = r'cat'
text = "The black cat sat on the mat."
search(exp, text)
```

```python
exp = r'cats'
text = "The black cat sat on the mat."
search(exp, text)
```

```python
exp = r'\d\d\d'
text = "Call me at 123-4567"
search(exp, text)
```

```python
exp = r'^[A-Z]\w+'
text = "Hello world"
search(exp, text)
```

```python
exp = r'python'
text = "I love JavaScript"
search(exp, text)
```

```python
exp = r'\d{4}-\d{2}-\d{2}'
text = "Today's date is 2025-11-05."
search(exp, text)
```

```python
def fullmatch(regular_expression, text):
    """this return result of fullmatch of regular-expressions"""
    result = re.fullmatch(regular_expression, text)
    if result:
        word = result.group()
        span = result.span()
        print(f'Word: {word}')
        print(f'Index location: {span}')
    else:
        print('No match found.')
```

```python
exp = r'cats'
text = "cats"
fullmatch(exp, text)
```

```python
exp = r'cats'
text = "black cat"
fullmatch(exp, text)
```

```python
exp = r'\d{3}'
text = "1234"
fullmatch(exp, text)
```

```python
exp = r'\w+@\w+\.\w+'
text = "user@example.com"
fullmatch(exp, text)
```

```python
exp = r'\d{4}-\d{2}-\d{2}'
text = "Date: 2025-11-05"
fullmatch(exp, text)
```

```python
def sub(regular_expression,replacement, text):
    """this return result of sub of regular-expressions"""
    result = re.sub(regular_expression, replacement, text)
    if result:
        print(f'{result}')
    else:
        print('Something went wrong.')
```

```python
exp = r'\d'
replacement = '#'
text = "My phone number is 12345."
sub(exp, replacement , text)
```

```python
exp = r'[^a-zA-Z0-9]'
replacement = ''
text = "Hello@2025!"
sub(exp, replacement , text)
```

```python
exp = r' +$'
replacement = ''
text = "Trailing spaces "
sub(exp, replacement , text)
```

```python
def split(delimeter, text):
    """this return result of split of regular-expressions"""
    result = re.split(delimeter, text)
    if result:
        print(f'{result}')
    else:
        print('Something went wrong.')
```

```python
exp = r'\s+'
text = "Hello World Python"
split(exp , text)
```

```python
exp = r'\d+'
text = "abc123def456ghi"
split(exp , text)
```

```python
exp = r'\s*,\s*'
text = "apple, banana , mango ,grape"
split(exp , text)
```

```python
def findall(regular_expression, text):
    """this return result of findall of regular-expressions"""
    result = re.findall(regular_expression, text)
    if result:
        for i, word in enumerate(result):
            print(f'{i+1}: {word}')
    else:
        print('Something went wrong.')
```

```python
exp = r'\d+'
text = "There are 3 cats, 14 dogs, and 1 parrot."
findall(exp , text)
```

```python
exp = r'[A-Z][a-z]+'
text = "Alice and Bob went to NewYork."
findall(exp , text)
```

```python
"""

8.23 (Project: Spam Scanner) Spam (or junk e-mail) costs U.S. organizations billions of dollars a year in spam-prevention 
software, equipment, network resources, bandwidth, and lost productivity. Research online some of the most common spam 
e-mail messages and words, and check your junk e-mail folder. Create a list of 30 words and phrases commonly found in spam
messages. Write an application in which the user enters an e-mail message. Then, scan the message for each of the 30 keywords
or phrases. For each occurrence of one of these within the message, add a point to the message’s “spam score.” Next, rate the
likelihood that the message is spam, based on the number of points it received. In the data science case study chapters, 
you’ll be able to attack this problem in a more sophis- ticated way.

"""
```

```python
spam_list = ['best price', 'bonus', 'cash', 'compare rates', 'confidentiality', 'double your', 'Earn', 'additional income',
            'bulk email', 'call', 'cancel any time', 'claims', 'cost', 'credit', 'limited time only', 'dear friend', 
             'fantastic', 'financial', 'hidden', 'increase sales', 'make money', 'be your own boss', '100% more', '100% free'
             , 'act now', 'earn money', 'fast cash', 'financial freedom', 'free access', 'free gift']
```

```python
import re
```

```python
def get_email():
    """get email from user"""
    while True:
        mail = input('Enter email: ').lower()
        if len(mail) > 0:
            return mail
        else:
            print('Please enter email.')
```

```python
def evaluate_spam(mail):
    """evaluate if email is spam or not """
    points = 0
    spam_words = []
    for word in spam_list:
        if word in mail:
            points += 1
            spam_words += [word]
    return points, spam_words
```

```python
def print_result(points, spam_words):
    """print result of spam detection"""
    if points <= 2:
        print('\nThis mail is not spam')
        return
    elif points <= 8:
        print(f'\nThis mail is spam with threat level "Low".\nSpam Score: {points}')
    elif points <= 12:
        print(f'\nThis mail is spam with threat level "Medium".\nSpam Score: {points}')
    else:
        print(f'\nThis mail is spam with threat level "High".\nSpam Score: {points}')
    print('Following is the list of words because of which this mail was marked as spam.')
    for i, word in enumerate(spam_words):
        print(f'{i+1}: {word}')
```

```python
mail = get_email()
points, spam_words = evaluate_spam(mail)
print_result(points, spam_words)
```

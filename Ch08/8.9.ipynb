```python
"""

8.9 (Creating Three-Letter Strings from a Five-Letter Word) Write a script that reads a five-letter word from the user and 
produces every possible three-letter string, based on the word’s letters. For example, the three-letter words produced from 
the word “bathe” include “ate,” “bat,” “bet,” “tab,” “hat,” “the” and “tea.” Challenge: Investigate the functions from the 
itertools module, then use an appropriate function to automate this task.

"""
```

```python
while True:
    user_in = input('Enter a five-letter word: ')
    if len(user_in) == 5:
        break
    else:
        print('Invalid input. The word must be of five-letters')
```

```python
counter = 1
for i in user_in:
    for j in user_in:
        if i == j:
            continue
        for k in user_in:
            if k == j or k == i:
                continue
            print(f'{counter}\t {i + j + k}')
            counter += 1
```

```python
import itertools
```

```python
l = itertools.permutations(user_in, 3)
```

```python
for index, value in enumerate(l):
    print(f'{index + 1}\t{''.join(value)}')
```

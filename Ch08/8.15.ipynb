```python
import re
```

```python
st = """
8.15 (Regular Expressions: Counting Characters and Words) Use regular expressions and the findall function to count the 
number of digits, non-digit characters, whitespace characters and words in a string.
"""
```

```python
digits = re.findall(r'\d', st)
```

```python
print(f'Number of digits characters are {len(digits)}.')
```

```python
non_digits = re.findall(r'\D', st)
```

```python
print(f'Number of non digits characters are {len(non_digits)}.')
```

```python
white_space = re.findall(r'\s', st)
```

```python
print(f'Number of white spaces are {len(white_space)}.')
```

```python
words = re.findall(r'\w+', st)
```

```python
print(f'Number of words are {len(words)}.')
```

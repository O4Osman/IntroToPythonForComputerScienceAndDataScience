```python
"""

5.29 (Intro to Data Science: Visualizing Survey Response Statistics) Using the list in Exercise 5.28 and the techniques 
you learned in Section 5.17.2, display a bar chart showing the response frequencies and their percentages of the total 
responses.

"""
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
ratings = [1, 2, 5, 4, 3, 5, 2, 1, 3, 3, 1, 4, 3, 3, 3, 2, 3, 3, 2, 5]
```

```python
values, frequencies = np.unique(ratings, return_counts=True)
```

```python
title = f'Food Ratings by {len(ratings)} Students'
```

```python
sns.set_style('whitegrid')
axes = sns.barplot(x=values, y=frequencies, palette='bright', hue=values)
axes.set_title(title)
axes.set(xlabel='Ratings', ylabel='Frequency')
axes.set_ylim(top=max(frequencies) * 1.10)
for bar, frequency in zip(axes.patches, frequencies):
    text_x = bar.get_x() + bar.get_width() / 2.0
    text_y = bar.get_height()
    text = f'{frequency:,}\n{frequency / len(ratings):.0%}'
    axes.text(text_x, text_y, text, fontsize=11, ha='center', va='bottom')
```

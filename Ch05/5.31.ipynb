```python
import matplotlib.pyplot as plt
import numpy as np
import random
import seaborn as sns
```

```python
# Flipping Coin 200 times
coin_flips = [random.randrange(1, 3) for i in range(200)]
values, frequencies = np.unique(coin_flips, return_counts=True)
title = f'Flipping a coin {len(coin_flips):,} Times'
sns.set_style('whitegrid')
axes = sns.barplot(x=values, y=frequencies, palette='bright', hue=values)
axes.set_title(title)
axes.set(xlabel='Coin Faces (1=Head, 2=Tail)', ylabel='Frequency')
axes.set_ylim(top=max(frequencies) * 1.10)
```

```python
# Flipping Coin 200 times with frequency and percentage
coin_flips = [random.randrange(1, 3) for i in range(200)]
values, frequencies = np.unique(coin_flips, return_counts=True)
title = f'Flipping a coin {len(coin_flips):,} Times'
sns.set_style('whitegrid')
axes = sns.barplot(x=values, y=frequencies, palette='bright', hue=values)
axes.set_title(title)
axes.set(xlabel='Coin Faces (1=Head, 2=Tail)', ylabel='Frequency')
axes.set_ylim(top=max(frequencies) * 1.10)
for bar, frequency in zip(axes.patches, frequencies):
    text_x = bar.get_x() + bar.get_width() / 2.0
    text_y = bar.get_height()
    text = f'{frequency:,}\n{frequency / len(coin_flips):.3%}'
    axes.text(text_x, text_y, text, fontsize=11, ha='center', va='bottom')
```

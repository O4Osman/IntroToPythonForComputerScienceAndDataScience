```python
import re
import itertools
```

```python
with open('words.txt', mode='r') as dictionary:
    words = []
    for word in dictionary:
        words.append(word.strip().lower())
```

```python
# words as string so that re module can search in words taking much less time
string_words = ' '.join(words)
```

```python
with open('test.txt', mode='r') as test:
    test_words = []
    for word in test:
        test_words.append(word.strip().lower())
```

```python
def check_transpositions(word):
    """it generates all possible combination of words from letters of passed parameter and find possible correct word"""
    patterns = []
    for i in range(len(word) - 1):
        n = word[i+1]
        c = word[i]
        patterns.append(''.join([word[:i], n, c, word[i+2:]]))

    for pattern in patterns:
        result =  re.search('\s'+pattern+'\s', string_words)
        if result:
            return result.group()
```

```python
def check_dubble_type(word):
    """it generate all the possbilities to check if user double type a letter and return correct word possible"""
    patterns = []
    for i in range(len(word) - 1):
        if word[i] == word[i+1]:
            patterns.append(''.join([word[:i+1] , word[i+2:]]))
    for pattern in patterns:
        result =  re.search('\s'+pattern+'\s', string_words)
        if result:
            return result.group()
```

```python
def check_wrong_typing(word):
    """it check user this word have one or wrong letter in the word"""
    patterns = []
    for index, letter in enumerate(word):
        if index != 0 and letter != ' ':
            patterns.append(word[:index] + '.' + word[index+1:])
    for pattern in patterns:
        result =  re.search('\s'+pattern+'\s', string_words)
        if result:
            return result.group()
```

```python
for word in test_words:
    if word not in words:
        print(f'Wrong Word: {word}')
        result = check_dubble_type(word)
        if result:
            print(f'Did you mean "{result.strip()}"')
        else:
            result1 = check_wrong_typing(word)
            if result1:
                print(f'Did you mean "{result1.strip()}"')
            else:
                result2 = check_transpositions(word)
                if result2:
                    print(f'Did you mean "{result2.strip()}"')
```

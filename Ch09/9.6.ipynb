```python
"""

9.6 (Class Average: Writing a Gradebook Dictionary to a JSON File) Reimplement Exercise 9.3 using the json module to write the student information to the file in JSON format. For this 
exercise, create a dictionary of student data in the following format:
gradebook_dict = {'students': [student1dictionary, student2dictionary, ...]}
Each dictionary in the list represents one student and contains the keys 'first_name', 'last_name', 'exam1', 'exam2' and 'exam3', which map to the values representing each studentâ€™s 
first name (string), last name (string) and three exam scores (integers). Output the gradebook_dict in JSON format to the file grades.json.

"""
```

```python
import json
```

```python
with open('grades.json', 'w') as grades:
    grades_list = []
    while True:
        try:  
            f_name = input('Enter first name: ')
            l_name = input('Enter last name: ')
            exam1 = int(input('Exam 1 grades: '))
            exam2 = int(input('Exam 2 grades: '))
            exam3 = int(input('Exam 3 grades: '))
        except ValueError:
            print('Invalid value. Please try again for this student"s record.')
        else:
            grade = {
                'first_name': f_name,
                'last_name': l_name,
                'exam1': exam1,
                'exam2': exam2,
                'exam3': exam3,
            }
            grades_list += [grade]
            con = input('Do you want to add more records? [Y/N]').lower()
            if con != 'y':
                break
    if len(grades_list) > 0:
        gradebook_dict = {
        'students': grades_list
        }
        json.dump(gradebook_dict, grades)
        print('All records have been saved.')
    else:
        print('No data to save.')
```

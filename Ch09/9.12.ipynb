```python
"""

9.12 (Project: State-of-the-Union Speeches) Text files of all U.S. Presidents’ State-ofthe-Union speeches are available online. Download one of these speeches. Write a script
that reads the speech from the file, then displays statistics about the speech, including the total word count, the total character count, the average word length, the average 
sentence length, a word distribution of the words frequencies, and the top 10 longest words. In the “Natural Language Processing (NLP)” chapter, you’ll find lots of more 
sophisticated techniques for analyzing and comparing such texts.

"""
```

```python
import re
from collections import Counter
```

```python
try:
    with open('Joe_Biden.txt', 'r') as speech:
        text = ''
        for line in speech:
            text += line
except:
    print('Some error occured during reading of file.')
```

```python
words = re.findall(r'\w+', text.lower())
```

```python
words_count = len(words)
```

```python
character_count = len(''.join(words))
```

```python
try:
    average_word_length = character_count/words_count
except ZeroDivisionError:
    print('Word Count is 0. Cannot find average word lenght')
```

```python
try:
    sentences = re.split(f'[.!?]+\s+|[.!?]+$', text)
    average_sentence_length = words_count/len(sentences)
except ZeroDivisionError:
    print('Sentence count is 0. Cannot find average sentence lenght')
```

```python
words_distribution = Counter(words)
```

```python
def check_len(word):
    return len(str(word))
top_10_longest_words = sorted(words, key=check_len, reverse=True)[:10]
```

```python
print(f'Total word count: {words_count}')
print(f'Total character count: {character_count}')
print(f'Average word length: {average_word_length:.2f}')
print(f'Average sentence length: {average_sentence_length:.2f}')

print(f'Word distribution of all words: ')
for word, count in sorted(words_distribution.items())[:4]:
    print(f'{word:<20}{count}')
print('....\n....\n....')
for word, count in sorted(words_distribution.items())[-4:]:
    print(f'{word:<20}{count}')

print('\n')
print(f'Top 10 longest words: ')
for i, word in enumerate(top_10_longest_words):
    print(f'{i+1}: {word:<20}  length: {len(str(word))}')
```

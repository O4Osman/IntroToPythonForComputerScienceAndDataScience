```python
"""

9.9 (Telephone-Number Word Generator) In Exercise 5.12, you created a telephone- number word-generator program. Modify that program to write its output to a text file.

"""
```

```python
numbers_table = [[],[],['A','B','C'], ['D', 'E', 'F'], ['G', 'H', 'I'], ['J', 'K', 'L'], 
                 ['M', 'N', 'O'], ['P', 'R', 'S'], ['T', 'U', 'V'], ['W', 'X', 'Y']]
```

```python
with open('telephone-number.txt', 'w') as tp_numbers:
    while True:
        try:
            user_input = input('Enter a 7 digit number (2-9): ')
            if len(user_input) == 7:
                digit_check = int(user_input)
                if 2000000 <= digit_check <= 9999999:
                    break
                else:
                    print('Invalid numbers. Valid numbers are 2 to 9.')
            else:
                print('Please enter a 7 digit number (2-9).')
        except ValueError:
            print('Please enter digits only (2-9).')
            
    dataset = []
    
    for i in user_input:
        dataset.append(numbers_table[int(i)])

    for i in dataset[0]:
        word = [''] * 7
        word[0]= i
        for j in dataset[1]:
            word[1] = j
            for k in dataset[2]:
                word[2] = k
                for l in dataset[3]:
                    word[3] = l
                    for m in dataset[4]:
                        word[4] = m
                        for n in dataset[5]:
                            word[5] = n
                            for o in dataset[6]:
                                word[6] = o
                                output = ''.join(word)
                                tp_numbers.write(output + '\n')
```

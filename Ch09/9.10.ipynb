```python
import re
from collections import Counter
```

```python
try:
    with open('pride.txt', 'r') as pride_and_prejudice:
        pride = ''
        for line in pride_and_prejudice:
            pride += line
except:
    print('Some error occured during reading of file.')
```

```python
words = re.findall(r'\w+', pride)
```

```python
words_count = len(words)
```

```python
character_count = len(''.join(words))
```

```python
try:
    average_word_length = character_count/words_count
except ZeroDivisionError:
    print('Word Count is 0. Cannot find average word lenght')
```

```python
try:
    sentences = re.split(f'[.!?]+\s+|[.!?]+$', pride)
    average_sentence_length = words_count/len(sentences)
except ZeroDivisionError:
    print('Sentence count is 0. Cannot find average sentence lenght')
```

```python
words_distribution = Counter(words)
```

```python
def check_len(word):
    return len(str(word))
top_10_longest_words = sorted(words, key=check_len, reverse=True)[:10]
```

```python
print(f'Total word count: {words_count}')
print(f'Total character count: {character_count}')
print(f'Average word length: {average_word_length:.2f}')
print(f'Average sentence length: {average_sentence_length:.2f}')

print(f'Word distribution of all words: ')
for word, count in sorted(words_distribution.items())[:4]:
    print(f'{word:<20}{count}')
print('....\n....\n....')
for word, count in sorted(words_distribution.items())[-4:]:
    print(f'{word:<20}{count}')

print('\n')
print(f'Top 10 longest words: ')
for i, word in enumerate(top_10_longest_words):
    print(f'{i+1}: {word:<20}  length: {len(str(word))}')
```

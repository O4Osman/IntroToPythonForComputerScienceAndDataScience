```python
"""

9.14 (Project: Basic Similarity Detection via Average Sentence Length and Average Word Length) Who actually wrote William Shakespeare’s works? Some researchers believe that Sir Francis Bacon may have 
authored some or all of these works. Download one of Shakespeare’s works and one of Bacon’s works from Project Gutenberg. For each, calculate the average sentence length and average word length. 
Are these close? Compute other statistics as well. 

"""
```

```python
import re
from collections import Counter
```

```python
def check_len(word):
    return len(str(word))
```

```python
def load_work(file_name):
    """load the text file into a variable and return that"""
    try:
        with open(file_name, 'r') as work:
            book = ''
            for line in work:
                book += line
    except:
        print('Some error occured during reading of file.')
    else:
        return book
```

```python
shakespeare = load_work('romeo_and_juliet.txt')
```

```python
bacon = load_work('new_atlantis.txt')
```

```python
# shapespeare
s_words = re.findall(r'\w+', shakespeare.lower())
s_words_count = len(s_words)
s_character_count = len(''.join(s_words))

try:
    s_average_word_length = s_character_count/s_words_count
except ZeroDivisionError:
    print('Word Count is 0. Cannot find average word lenght')

try:
    s_sentences = re.split(f'[.!?]+\s+|[.!?]+$', shakespeare)
    s_average_sentence_length = s_words_count/len(s_sentences)
except ZeroDivisionError:
    print('Sentence count is 0. Cannot find average sentence lenght')

s_words_distribution = Counter(s_words)

s_top_10_longest_words = sorted(s_words, key=check_len, reverse=True)[:10]
print(f'Total word count: {s_words_count}')
print(f'Total character count: {s_character_count}')
print(f'Average word length: {s_average_word_length:.2f}')
print(f'Average sentence length: {s_average_sentence_length:.2f}')

print(f'Word distribution of all words: ')
for word, count in sorted(s_words_distribution.items())[:4]:
    print(f'{word:<20}{count}')
print('....\n....\n....')
for word, count in sorted(s_words_distribution.items())[-4:]:
    print(f'{word:<20}{count}')

print('\n')
print(f'Top 10 longest words: ')
for i, word in enumerate(s_top_10_longest_words):
    print(f'{i+1}: {word:<20}  length: {len(str(word))}')
```

```python
# Bacon
b_words = re.findall(r'\w+', bacon.lower())
b_words_count = len(b_words)
b_character_count = len(''.join(b_words))

try:
    b_average_word_length = b_character_count/b_words_count
except ZeroDivisionError:
    print('Word Count is 0. Cannot find average word lenght')

try:
    b_sentences = re.split(f'[.!?]+\s+|[.!?]+$', bacon)
    b_average_sentence_length = b_words_count/len(b_sentences)
except ZeroDivisionError:
    print('Sentence count is 0. Cannot find average sentence lenght')

b_words_distribution = Counter(b_words)

b_top_10_longest_words = sorted(b_words, key=check_len, reverse=True)[:10]
print(f'Total word count: {b_words_count}')
print(f'Total character count: {b_character_count}')
print(f'Average word length: {b_average_word_length:.2f}')
print(f'Average sentence length: {b_average_sentence_length:.2f}')

print(f'Word distribution of all words: ')
for word, count in sorted(b_words_distribution.items())[:4]:
    print(f'{word:<20}{count}')
print('....\n....\n....')
for word, count in sorted(b_words_distribution.items())[-4:]:
    print(f'{word:<20}{count}')

print('\n')
print(f'Top 10 longest words: ')
for i, word in enumerate(b_top_10_longest_words):
    print(f'{i+1}: {word:<20}  length: {len(str(word))}')
```

```python
if (b_average_word_length - 1) <= s_average_word_length <= (b_average_word_length + 1):  # creating a little bit of margin
    print('According to word lenght both work belongs to one person. As word length for both is close.')
else:
    print('According to word lenght both work belongs to different person. As word length for both is not close.')
```

```python
if (b_average_sentence_length - 1) <= s_average_sentence_length <= (b_average_sentence_length + 1):  # creating a little bit of margin
    print('According to sentence lenght both work belongs to one person. As sentence length for both is close.')
else:
    print('According to sentence lenght both work belongs to different person. As sentence length for both is not close.')
```

```python
"""

9.8 (pickle Object Serialization and Deserialization) We mentioned that we prefer to use JSON for object serialization due to the Python documentation’s stern security warnings about 
pickle. However, pickle has been used to serialize objects for many years, so you’re likely to encounter it in Python legacy code. According to the documentation, “If you are doing your 
own pickle writing and reading, you’re safe (provided no one else has access to the pickle file, of course.)”14 Reimplement your solutions to Exercises 9.6– 9.7 using the pickle module’s
dump function to serialize the dictionary into a file and its load function to deserialize the object. Pickle is a binary format, so this exercise requires binaryfiles. 
Use the file-open mode 'wb' to open the binary file for writing and 'rb' to open the binary file for reading. Function dump receives as arguments an object to serialize and a file object in which to 
write the serialized object. Function load receives the file ob- ject containing the serialized data and returns the original object. The Python documentation suggests the pickle file
extension .p.

"""
```

```python
import pickle
```

```python
with open('grades.p', 'wb') as grades:
    grades_list = []
    while True:
        try:  
            f_name = input('Enter first name: ')
            l_name = input('Enter last name: ')
            exam1 = int(input('Exam 1 grades: '))
            exam2 = int(input('Exam 2 grades: '))
            exam3 = int(input('Exam 3 grades: '))
        except ValueError:
            print('Invalid value. Please try again for this student"s record.')
        else:
            grade = {
                'first_name': f_name,
                'last_name': l_name,
                'exam1': exam1,
                'exam2': exam2,
                'exam3': exam3,
            }
            grades_list += [grade]
            con = input('Do you want to add more records? [Y/N]').lower()
            if con != 'y':
                break
    if len(grades_list) > 0:
        gradebook_dict = {
        'students': grades_list
        }
        pickle.dump(gradebook_dict, grades)
        print('All records have been saved.')
    else:
        print('No data to save.')
```

```python
with open('grades.p', 'rb') as grades:
    grades_dict = pickle.load(grades)['students']
    print(f'{"First Name":<10}\t{"Last Name":<10}\t{"Exam 1 Grade":>12}\t{"Exam 2 Grade":>12}\t{"Exam 3 Grade":>12}\t{"Student Average":>12}')
    exam1_total = 0
    exam2_total = 0
    exam3_total = 0
    student_count = 0
    for grade in grades_dict:
        first_name = grade['first_name']
        last_name = grade['last_name']
        exam1 = grade['exam1']
        exam2 = grade['exam2']
        exam3 = grade['exam3']
        exam1_total += exam1
        exam2_total += exam2
        exam3_total += exam3
        student_count += 1
        average = sum([exam1, exam2, exam3]) / 3
        print(f'{first_name:<10}\t{last_name:<10}\t{exam1:>12}\t{exam2:>12}\t{exam3:>12}\t{average:>12.2f}')
    if student_count > 0:
        avg1 = exam1_total / student_count
        avg2 = exam2_total / student_count
        avg3 = exam3_total / student_count
        print(f'{"Class Average":24}\t{avg1:>12.2f}\t{avg2:>12.2f}\t{avg3:>12.2f}')
    else:
        print('No students record found to calculate class average exam wise.')
```

```python
"""

9.4 (Class Average: Reading Student Records from a CSV File) Use the csv module to read the grades.csv file from the previous exercise. Display the data in tabular format.

"""
```

```python
import csv
```

```python
with open('grades.csv', 'r') as grades:
    reader = csv.reader(grades)
    print(f'{"First Name":<10}\t{"Last Name":<10}\t{"Exam 1 Grade":>12}\t{"Exam 2 Grade":>12}\t{"Exam 3 Grade":>12}')
    for record in reader:
        f_name, l_name, exam1, exam2, exam3 = record
        print(f'{f_name:<10}\t{l_name:<10}\t{exam1:>12}\t{exam2:>12}\t{exam3:>12}')
```

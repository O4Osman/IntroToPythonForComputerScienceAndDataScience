```python
from wordcloud import WordCloud
import re
from collections import Counter
```

```python
wordcloud = WordCloud(colormap='viridis', background_color='white') # prism was giving error
```

```python
# Windows opens files with the cp1252 encoding by default, but file pride.txt contains characters that cp1252 cannot decode.
try:
    with open('pride.txt', 'r',  encoding='utf-8') as pride_and_prejudice: 
        pride = ''
        for line in pride_and_prejudice:
            pride += line
except:
    print('Some error occured during reading of file.')
```

```python
words = re.findall(r'\w+', pride)
```

```python
words_distribution = Counter(words)
```

```python
sorted_words_distribution = sorted(words_distribution.items(), key=lambda item: item[1], reverse=True)[:200]
```

```python
top_200_words_distribution = {key: value for key, value in sorted_words_distribution}
```

```python
wordcloud = wordcloud.fit_words(top_200_words_distribution)
wordcloud.to_file('PrideAndPrejudice.png')
```

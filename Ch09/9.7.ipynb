```python
"""

9.7 (Class Average: Reading a Gradebook Dictionary from a JSON File) Reimplement Exercise 9.4 using the json module to read the grades.json file created in the previous exercise. 
Display the data in tabular format, including an additional column showing each student’s average to the right of that student’s three exam grades and an additional row showing the class
average on each exam below that exam’s column.

"""
```

```python
import json
```

```python
with open('grades.json', 'r') as grades:
    grades_dict = json.load(grades)['students']
    print(f'{"First Name":<10}\t{"Last Name":<10}\t{"Exam 1 Grade":>12}\t{"Exam 2 Grade":>12}\t{"Exam 3 Grade":>12}\t{"Student Average":>12}')
    exam1_total = 0
    exam2_total = 0
    exam3_total = 0
    student_count = 0
    for grade in grades_dict:
        first_name = grade['first_name']
        last_name = grade['last_name']
        exam1 = grade['exam1']
        exam2 = grade['exam2']
        exam3 = grade['exam3']
        exam1_total += exam1
        exam2_total += exam2
        exam3_total += exam3
        student_count += 1
        average = sum([exam1, exam2, exam3]) / 3
        print(f'{first_name:<10}\t{last_name:<10}\t{exam1:>12}\t{exam2:>12}\t{exam3:>12}\t{average:>12.2f}')
    if student_count > 0:
        avg1 = exam1_total / student_count
        avg2 = exam2_total / student_count
        avg3 = exam3_total / student_count
        print(f'{"Class Average":24}\t{avg1:>12.2f}\t{avg2:>12.2f}\t{avg3:>12.2f}')
    else:
        print('No students record found to calculate class average exam wise.')
```

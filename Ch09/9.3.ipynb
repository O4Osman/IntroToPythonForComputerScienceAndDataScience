```python
"""

9.3 (Class Average: Writing Student Records to a CSV File) An instructor teaches a class in which each student takes three exams. The instructor would like to store this information 
in a file named grades.csv for later use. Write code that enables an instructor to enter each student’s first name and last name as strings and the student’s three exam grades as 
integers. Use the csv module to write each record into the grades.csv file. Each record should be a single line of text in the following CSV format:
firstname,lastname,exam1 grade,exam2 grade,exam3 grade

"""
```

```python
import csv
```

```python
with open('grades.csv', 'w', newline='') as grades:
    writer = csv.writer(grades)
    while True:
        try:  
            f_name = input('Enter first name: ')
            l_name = input('Enter last name: ')
            exam1 = int(input('Exam 1 grades: '))
            exam2 = int(input('Exam 2 grades: '))
            exam3 = int(input('Exam 3 grades: '))
        except ValueError:
            print('Invalid value. Please try again for this student"s record.')
        else:
            writer.writerow([f_name, l_name, exam1, exam2, exam3])
            con = input('Do you want to add more records? [Y/N]').lower()
            if con != 'y':
                break
    print('All records have been saved.')
```

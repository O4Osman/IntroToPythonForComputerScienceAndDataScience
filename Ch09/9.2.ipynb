```python
"""

9.2 (Class Average: Reading Grades from a Plain Text File) In an IPython session, write code that reads the grades from the grades.txt file you created in the previous exercise. 
Display the individual grades and their total, count and average.

"""
```

```python
with open('grades.txt', 'r') as grades:
    grade_list = []
    for index, grade in enumerate(grades):
        grade_list.append(int(grade))
        print(f'{index + 1}: {grade}', end='')
    total = sum(grade_list)
    count = len(grade_list)
    print(f'Total: {total}')
    print(f'Count: {count}')
    print(f'Average: {total/count:.2f}')
```

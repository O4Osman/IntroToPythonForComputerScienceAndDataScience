```python
from decimal import Decimal
```

```python
# After searching from multiple sources I will use the following data available on 16-9-2025
current_population = Decimal('8246688341')
growth_rate = 0.85 / 100
growth_rate = Decimal(growth_rate)
```

```python
print(f'Year\tWorld Population\tNumerical Increment')

for year in range (1, 101):
    numerical_increment = Decimal(current_population * growth_rate)
    current_population += numerical_increment
    print(f'{year:>3}\t{int(current_population)}\t\t{int(numerical_increment)}')
```
